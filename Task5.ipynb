```python
# Problem 1: Post Engagement Analyzer

def analyze_engagement(likes_list):
    total_likes = 0

    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like

    # Conditional check
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

# Example Usage
likes = [200, 300, 150, 400]
analyze_engagement(likes)
```

```python
# Problem 2: Medicine Stock Alert System

def check_medicine_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)

# Example Usage
check_medicine_stock(6)
```

```python
# Problem 3: Rainfall Adequacy Checker

def rainfall_checker(rainfall_data, required_level):
    total = 0

    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

# Example Usage
rainfall = [70, 75, 80, 65]
rainfall_checker(rainfall, 72)
```

```python
# Problem 4: Duplicate Account Detection

def detect_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)

# Example Usage
users = ["ram", "sita", "ram", "john"]
detect_duplicates(users)
```

```python
# Problem 5: Appointment Eligibility Checker

def check_appointment_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)

# Example Usage
check_appointment_eligibility(21)
```

```python
# Problem 6: Premium Crop Price Filter

def filter_premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

# Example Usage
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

```python
# Problem 7: Application Health Checker

def check_system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

# Example Usage
check_system_health(7)
```

```python
# Problem 8: Daily Transaction Limit Checker

def transaction_checker(amount):
    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

# Example Usage
transaction_checker(60000)
```

```python
# Problem 9: Attendance Eligibility System

def attendance_checker(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for day in attendance_list:
        if day == 1:  # 1 means present
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

# Example Usage
attendance = [1,1,1,0,1,1,1,1,0,1]
attendance_checker(attendance)
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def electricity_bill(units):
    bill = 0

    # Billing calculation using slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Classification:", usage)

# Example Usage
electricity_bill(250)
```
