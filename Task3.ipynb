# Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Finding the top score
top_score = max(employees.values())

top_members = []
for name, score in employees.items():
    if score == top_score:
        top_members.append(name)

# Joining names and result output
result = ", ".join(top_members)
print(f"Top Performers Eligible for Bonus: {result} (Score: {top_score})")
```

# Search Query Keyword Analysis

```python
# Problem 2: Search Query Keyword Analysis

import re

query = "Buy mobile phone buy phone online"

# Convert the input to lowercase
cleaned = re.sub(r'[^\w\s]', '', query.lower())

# Split into individual words
words = cleaned.split()

# Count the frequency of each keyword
freq = {}
for word in words:
    if word in freq:
        freq[word] = freq[word] + 1
    else:
        freq[word] = 1

# Storing values which can appear more
result = {}
for word in freq:
    if freq[word] > 1:
        result[word] = freq[word]

print(result)
```

# Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for i, val in enumerate(sensor_readings):
    # Identify readings that are even numbers (valid readings)
    if val % 2 == 0:
    #Store them as (hour_index, reading_value) pair
        valid_readings.append((i, val))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}
for email in emails:
    domain = email.split('@')[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)
for domain, count in domain_count.items():
    print(f"{domain}: {int(count / total * 100)}%")
```

# Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

average = sum(sales) / len(sales)   #Calculate the average daily sales
threshold = average * 1.30         #Detect days where sales are more than 30% above averag

for i, sale in enumerate(sales, start=1):
    if sale > threshold:
        print(f"Day {i}: {sale}")     #Display the day number and sale value
```

# Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

freq = {}
for uid in user_ids:
    freq[uid] = freq.get(uid, 0) + 1         #Identify duplicate user IDs

for uid, count in freq.items():
    if count > 1:
        print(f"{uid} → {count} times")      #Display how many times each duplicate appears
```
